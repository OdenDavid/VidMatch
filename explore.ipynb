Download Thumbnails

```python
import os
import requests

# Function to download YouTube video thumbnails
def download_thumbnail(video_id, save_folder):
    thumbnail_url = f"https://img.youtube.com/vi/{video_id}/0.jpg"
    thumbnail_path = os.path.join(save_folder, f"{video_id}.jpg")
    response = requests.get(thumbnail_url)
    
    # Check if the file already exists
    if os.path.exists(thumbnail_path):
        return
    
    if response.status_code == 200:
        # Create the save folder if it doesn't exist
        if not os.path.exists(save_folder):
            os.makedirs(save_folder)

        # Save the thumbnail image to the specified folder
        with open(thumbnail_path, 'wb') as file:
            file.write(response.content)

        return thumbnail_path
    else:
        print(f"Failed to download thumbnail for video ID {video_id}. Status code: {response.status_code}")
        return "Not found"

# Example usage
save_folder = "../thumbnails/"  # Folder name

for id in final_sample["video_id"]:
    value = download_thumbnail(id, save_folder)
    if value == "Not found":
        final_sample = final_sample[final_sample['video_id'] != id]
```
